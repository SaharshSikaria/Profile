## Train-Test Split

```python
from sklearn.model_selection import train_test_split
X = df.drop('SalePrice',axis = 1).copy()
y = np.log1p(df['SalePrice']).copy()

X_train,X_valid,y_train,y_valid = train_test_split(X,y,test_size=0.2,random_state=0)
```

**As discussed above, the target value is converted to a log scale.**

## Dealing with Missing and Categorical values along with log transformation

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
transformer = FunctionTransformer(np.log1p)
```

```python
num_categorical_transformer = SimpleImputer(strategy='most_frequent')
num_noncategorical_transformer_simple = SimpleImputer()
num_noncategorical_transformer_log = Pipeline(steps = [
    ('imputer' , SimpleImputer()),
    ('logtransformation' , transformer)
])
qual_categorical_transformer_ordinal = Pipeline(steps = [
    ('imputer' , SimpleImputer(strategy='most_frequent')),
    ('ordinal' , OrdinalEncoder(handle_unknown = 'use_encoded_value',unknown_value= 42) , )
    ])
qual_categorical_transformer_OneHot = Pipeline(steps = [
    ('imputer' , SimpleImputer(strategy='most_frequent')),
    ('OHE' , OneHotEncoder(handle_unknown = 'ignore'))
    ])
```

```python
qual_categorical_ordinal = [col for col in qual_categorical if X[col].nunique()>=7]
qual_categorical_OneHot = [col for col in qual_categorical if X[col].nunique()<7]
Log_Transform_lst = '1stFlrSF GrLivArea LotArea'.split()
num_categorical = [x for x in num_categorical if x not in Log_Transform_lst]
```

```python
preprocessor = ColumnTransformer (transformers=[
    ('numcat' , num_categorical_transformer,num_categorical),
    ('numnoncatsimple' , num_noncategorical_transformer_simple,num_noncategorical),
    ('numnoncatlog' , num_noncategorical_transformer_log , Log_Transform_lst),
    ('qualcatord' , qual_categorical_transformer_ordinal,qual_categorical_ordinal),
    ('qualcatOHE' , qual_categorical_transformer_OneHot,qual_categorical_OneHot)
    ])
```

## Implementing Model and Pipeline

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

model = LinearRegression()
my_pipeline = Pipeline(steps = [('prepocessor' , preprocessor),
                                ('model' , model)
                               ])
my_pipeline.fit(X_train,y_train)
preds = my_pipeline.predict(X_valid)
score = np.sqrt(mean_squared_error(y_valid, preds))
print('RSME:', score)
print('MAE:' , mean_absolute_error(np.exp(y_valid),np.exp(preds)))
```

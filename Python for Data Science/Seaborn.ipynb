# Categorical Plots

## Bar Plots

One of the most basic categorical plot is the bar plot. Let's see an example

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

The barplot shows an aggregate value of numerical variable for each group of the categorical value passed. One variable passed must be categorical and the other must be numerical. When it comes to Categorical variable, the variable can be distinct natural also, like size in the above data. parameters like hue can also be passed to represent two categories.

The default estimator/aggregate value that the bar plot shows is the average or mean. However it can be changed by using an estimator parameter in the function call 

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

This plot returns standard deviation for each group in the categorical variable

## Count Plots

This essentially is also a barplot but with number of occurance of each variable in the category.

```python
sns.countplot(x='sex',data=tips)
```

## Box Plot

Boxplot is a very powerful plot as it shows all the quartile ranges in the data. It also Helps compare categories against one another. 

Boxplot also takes in one categorical variablee and one numerical variable. 

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

The bottom most line is the minimum value. The lower line of the box is the 25th Percentile or the 1st quartile. The middle line is the median and the topmost line of the box is the 3rd quartile. the topmost line of each plot is the maximum value. The box shows the quartile range and the wishkers shows the rest of the value from max to min.

The points above the plot is the outliers according to the data

### Adding a hue parameter

hue parameter takes in another categorical value. It further splits the initial categories depending on the categorical value passed in the hue parameter. This makes seaborn very powerful.

Let's see an example

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')

# This gives us further leavrage to analyse the primary category depending on a further category 
```

## Violin Plot

This is very similar to boxplot. However it gives us much more information. 

```python
sns.boxplot(x='day',y='total_bill',data=tips) #for comparison
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

The violin plot helps us to plot all the components that corresponds to actual datapoints. If we split the plots, it is the groupwise distribution of the data on the side. Box Plots are much easier to decipher though. 

We can also use hue functionality on the violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

Since, the violin plot is showing the information two times, we can essentially improve it by adding the split argument

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

## Strip plot

This is essentially a scatter plot where one variable is categorical.

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False) 

# generally keep the jitter True which is now by default. It's been negated to show how this is similar to pair plot
```

The above graph is very difficult to read in the sense that it doesn't give us give us the good sense for the density of the points.

TO improve it, we can use the jitter parameter

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)

# By default, jitter is true only.
```

```python
# We can also add hue to this.

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# Matrix Plots

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data (later in the machine learning section we will learn how to formally cluster data).

Let's begin by exploring seaborn's heatmap and clutermap:

```python
# Importing a new dataset

flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.describe(include='all')
```

## Heat Maps

Heat Maps are the way to print a graphical matrix and encode it with colour.

In order for heatmaps to work the data passed should be in matrix form.

Also, only data with numerical values should be passed into heatmaps.

```python
sns.heatmap(tips)

# This returns a value error as data contains columns with type other than numerics.
```

```python
sns.heatmap(tips[['total_bill','tip','size']])

# This does not make sense. 
# This is because the index values of the data was not representing anything. they were just count
# For heat maps to work the data must be in a matrix form. Let's explore the ways to convert data to matrix
```

### Correlation Matrix

```python
tips.corr()

# Here both index and columns represent a meaning in relation to each other
```

```python
sns.heatmap(tips.corr())
```

This plot makes sense. This is a very good visual way of analysing the data. As it will show the co-relation between the target variable and all the features we are considering. 

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot = True)

# cmap helps change the color scheme of the heatmap
# annot helps us print values on the map for easy comparisons
```

### Pivot Table

Another way of creating matrix form of data is using pivot table.

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
# Let's make a heatmap out of this value

pivotflights = flights.pivot_table(values='passengers',index='month',columns='year')
sns.heatmap(pivotflights)
```

```python
# To personalise these heatmaps

sns.heatmap(pivotflights,cmap='coolwarm',lw=2,linecolor='black')
```

```python
sns.heatmap(pivotflights,cmap='magma',lw=1,linecolor='white')
```

## Cluster Maps

These are heatmaps only. However, they don't represent the data in order. They perform hierarchical clustering on the data and produce a clustered version of the heatmap.

```python
sns.clustermap(pivotflights)
```

Notice now how the years and months are no longer in order, instead they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)

```python
sns.clustermap(pivotflights,cmap = 'coolwarm',standard_scale=1)
```

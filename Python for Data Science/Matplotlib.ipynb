# Matplotlib Object Oriented Method

This method gives us much greated control and many option to personalise the plots. Let's start with a formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object to bring changes to it.

## Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create a figure. This will be an empty canvas
fig = plt.figure()

#The first step is to add an axis to the blank canvas. 
axis = fig.add_axes([0.1,0.1,0.8,0.8]) #Left, Bottom, width, height (range of each parameter from 0 to 1)

# The first two parameters helps us specify the the position of origin wrt the left-bottom 
# corner of the canvas. The last two parameteras help us to decide the size of the axis box

axis.plot(x,y,'b')
axis.set_xlabel('This is X label') # we use the same commands as before just with addition of set_
axis.set_ylabel('This is Y label')
axis.set_title('This is the title')
```

The above approach seems a little worse for now but eventually we will see how it provides us with much better control

### Let's try an example which would have been very difficult with the functional approach

In this example we will plot a small graph inside the axes of another plot

```python
# Creating a blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
# This will be our larger figure
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # The percentages are always wrt to main canvas
# This will be our smaller figure

#Lets define the larger figure
axes1.plot(x,y,'r')
axes1.set_xlabel('X label of large plot')
axes1.set_ylabel('Y label of large plot')
axes1.set_title('Title of the large plot')

#Now for the smaller plot
axes2.plot(y,x,'b')
axes2.set_xlabel('X label of small plot')
axes2.set_ylabel('Y label of small plot')
axes2.set_title('Title of the small plot')

plt.show()
```

### Creating Subplots like this can be hard. Let's explore an easier way of creating Multiplots

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
#we will use plt.subplot(). plt.figure()returned a single figure object
#plt.subplot returns a 2 object tuple where the first element is the figure object
# the second is an array of axes object

fig, axes = plt.subplots()

#let's create a single plot for once
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

The Y2 label is overlaping. This can be a result in various scenarios

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
fig,axes = plt.subplots(1,2)

# we can also use a loop to iterate over axes array
for ax in axes:
    ax.plot(x,y)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
```

```python
fig.tight_layout() # Magic Command!!

fig
```

## Ploting multiple curves on the same axes

It is fairly easy to do this in Matplotlib. Any number of plots can be plotted on the same axes. just call the plot function on the data

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.plot(y,x)
axes.plot(range(0,25),range(0,25))
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

### Adding Legends to figures with multiple plotlines

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y,label = 'x**2')
axes.plot(y,x,label = 'x**1/2')
axes.plot(range(0,25),range(0,25),label='y=x')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
axes.legend(loc = 0)
```

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig,axes = plt.subplots(nrows=1,ncols =3,figsize=(12,3))

axes[0].plot(x,y)
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].set_title('Title 1: Default')


axes[1].plot(x,y)
axes[1].axis('tight')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].set_title('Title 2')

axes[2].plot(x,y)
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].set_ylim([0,30]) # These two are the new commands
axes[2].set_xlim([2,6])  #
axes[2].set_title('Title 3')
```

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with MatLab like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

### Colors with the color= parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Line and marker styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
